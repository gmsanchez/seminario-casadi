## Diferenciación automática

Consideremos:

\begin{equation}
\dot{p} = (1 - q^2)p-q+u
\end{equation}
\begin{equation}
\dot{q} = p
\end{equation}
\begin{equation}
\dot{c} = p^2+q^2+u^2
\end{equation}

```python
t  = SX.sym("t")        # time
u  = SX.sym("u")        # control
x  = SX.sym("[p,q,c]")  # state
p,q,c = x
```

```python
ode = vertcat([(1 - q**2)*p - q + u, p, p**2+q**2+u**2])
print ode, ode.shape
```

```python
J = jacobian(ode,x)
print J
```

```python
f = SXFunction("f",[t,u,x],[ode])

ffwd = f.derForward(1)

fadj = f.derReverse(1)

# side-by-side printing
print '{:*^24} || {:*^28} || {:*^28}'.format("f","ffwd","fadj")
def short(f):
    import re
    return re.sub(r", a\.k\.a\. \"(\w+)\"",r". \1",str(f).replace(", No description available","").replace("Input ","").replace("Output ",""))
for l in zip(short(f).split("\n"),short(ffwd).split("\n"),short(fadj).split("\n")):
  print '{:<24} || {:<28} || {:<28}'.format(*l)
```

Performing **forward sweeps** gives the **columns** of J

```python
print I
```

```python
for i in range(3):
  print ffwd([ t,u,x, ode, 0,0,I[:,i]  ])[0]
```

```python
print J
```

Performing **adjoint sweeps** gives the **rows** of J

```python
for i in range(3):
  print fadj([ t,u,x, ode, I[:,i]  ])[2]
```

## Integradores

$\dot{x}=f(x,u,t)$ with $x = [p,q,c]^T$

```python
f = SXFunction("f",daeIn(x=x,t=t,p=u),daeOut(ode=ode))

f.printDimensions()
```

Construye una función de integración $x_{k+1} = \Phi(f;\Delta t;x_k,u_k)$

```python
tf = 10.0
N = 20
dt = tf/N
```

```python
Phi = Integrator("Phi","cvodes",f,{"tf":dt})

x0 = vertcat([0,1,0])

print Phi({"x0":x0})
```

```python
x = x0
xs = [x]

for i in range(N):
  x = Phi({"x0":x})["xf"]
  
  xs.append(x)
```

```python
plot(horzcat(xs).T)
legend(["p","q","c"])
```

## Expresiones escalares (MX)

> Note 4: this is what makes CasADi stand out among AD tools

Recordemos

```python
n = 3

A = SX.sym("A",n,n)
B = SX.sym("B",n,n)
C = mul(A,B)
print C
dotdraw(C,direction='BT')
```

¿Y si **no queremos** expandir el producto en operaciones escalares?  (evitamos almacenar $O(n^3)$)

```python
A = MX.sym("A",n,n)
B = MX.sym("B",n,n)
C = mul(A,B)
print C
dotdraw(C,direction='BT')
```

What if you **cannot** expand into matrix operations?  ( numerical algorithm )

```python
C = solve(A,B)
print C
dotdraw(C,direction='BT')
```

```python
X0 = MX.sym("x",3)

XF = Phi({"x0":X0})["xf"]
print XF
```

```python
expr = sin(XF)+X0
dotdraw(expr,direction='BT')
```

## Functions of MX graphs

```python
F = MXFunction("F",[X0],[  expr  ])
print F
```

```python
print F([x0])
```

```python
J = F.jacobian()

print J([ x0 ])
```

This shows how an integrator-call can be embedded in matrix graph.
